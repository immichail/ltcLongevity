```python
import requests
import pandas as pd

def getRecs(userId: int, active = 1):
    url = f'http://localhost:11059/search?userId={userId}'
    if active is not None:
        url += f'&active={active}'
    
    try:
        res = requests.get(url)
        res = res.json()
        print(url)
    except:
        url = f'http://localhost:11059/search?userId={userId}'
        res = requests.get(url)
        res = res.json()
        print(url)
        
    return list(set([i['groupId'] for i in res['searchActivities']]))

df = pd.read_csv('./data/test(1).csv')
df['уникальный номер группы'] = df['уникальный номер участника'].apply(lambda x: getRecs(int(x), active = 1))
df.to_csv('./out/test.onlyActive.csv')
df['уникальный номер группы'] = df['уникальный номер участника'].apply(lambda x: getRecs(int(x), active = None))
df.to_csv('./out/test.all.csv')
```
