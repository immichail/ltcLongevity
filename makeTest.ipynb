```python
import requests
import pandas as pd

def getRecs(userId: int):
    print(userId)
    res = requests.get(f'http://localhost:11059/search?userId={userId}')
    return list(set([i['groupId'] for i in res.json()['searchActivities']]))

df = pd.read_csv('./data/test(1).csv')
df['уникальный номер группы'] = df['уникальный номер участника'].apply(lambda x: getRecs(int(x)))
```

```python
import matplotlib.pyplot as plt

plt.hist(df['уникальный номер группы'].apply(lambda x: len(x)), bins = 20)
```
